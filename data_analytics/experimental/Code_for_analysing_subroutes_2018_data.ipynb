```python
f = open('/Users/laura/Desktop/Trimester_3/research-project/web_app/journeyplanner/static/journeyplanner/ordered_stops_main.json',) 
ordered_stop_data = json.load(f)
```

```python
routes = list(ordered_stop_data.keys())
```

```python
directions = ["1", "2"]
```

```python
main_dict = {}
sub_dict = {}
```

```python
for key, value in ordered_stop_data.items():
            print(key)
            main_dict[key]= {}
            sub_dict[key] = {}
            main_dict[key]["1"] = {}
            sub_dict[key]["1"] = {}
            main_dict[key]["2"] = {}
            sub_dict[key]["2"] = {}
                
            ordered_stop_data_route = ordered_stop_data[str(key)]
        
            for key1, value1 in ordered_stop_data_route.items():
                print(key1)
                if ordered_stop_data_route[key1]['direction'] == 1:
                    
                    if ordered_stop_data_route[key1]['main'] == True:
                        main_subroute_1 = ordered_stop_data_route[key1]['stops']
                        main_dict[key]["1"][key1] = main_subroute_1
                    else:
                        subroute_1 = ordered_stop_data_route[key1]['stops']
                        sub_dict[key]["1"][key1] = subroute_1
                        
                elif ordered_stop_data_route[key1]['direction'] == 2:
                    
                    if ordered_stop_data_route[key1]['main'] == True:
                        main_subroute_2 = ordered_stop_data_route[key1]['stops']
                        main_dict[key]["2"][key1] = main_subroute_2
                    else:
                        subroute_2 = ordered_stop_data_route[key1]['stops']
                        sub_dict[key]["2"][key1] = subroute_2

print('MAIN DICT')                        
print(main_dict)
print()
print('SUB DICT')
print(sub_dict)
                    
                    
                
```

```python
keep_main = []
delete = []
manage = []
problematic_subroute_stops = []
for route in routes:

    for direction in directions:

        main_list_dict = main_dict[route][direction]
        sub_list_dict = sub_dict
        
        for key, value in main_list_dict.items():
            main_list = value
            keep_main.append(key)

            
            for key_sub, value_sub in sub_list_dict[route][direction].items():
                list_of_annoying_stops = list(np.setdiff1d(value_sub, main_list))
                if len(list_of_annoying_stops) == 0:
                    delete.append(key_sub)
                else:
                    manage.append(key_sub)
                    entry = [key_sub]
                    entry.append(list_of_annoying_stops)
                    problematic_subroute_stops.append(entry)
                    #print(key_sub, list_of_annoying_stops)
                    #problematic_stops[route]= {}
                    #problematic_stops[route][direction]= {}
                    #problematic_stops[route][direction][key_sub] = list_of_annoying_stops
print("MAIN")
print(keep_main)
print(len(keep_main))
print()
print("DELETE")
print(delete)
print(len(delete))
print()
print("MANAGE")
print(manage)
print(len(manage))
print()
print("PROBLEMATIC STOPS")
print(problematic_subroute_stops)
```
