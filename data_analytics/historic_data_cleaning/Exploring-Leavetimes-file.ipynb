```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('split_leave_times/leave_times_53.csv', keep_default_na=True, sep=',\s+', delimiter=';', skipinitialspace=True)
```

```python
df.head()
```

,DATASOURCE - constant - delete
DAYOFSERVICE = convert to datetime
TRIPID - convert to category
PROGNUMBER = sequential position of the stop point in the trip - covert to category
STOPPOINTID - convert to category
PLANNEDTIME_ARR
PLANNEDTIME_DEP
ACTUALTIME_ARR
ACTUALTIME_DEP
VEHICLEID - category
PASSENGERS - empty - delete
PASSENGERSIN - empty - delete
PASSENEGERSOUT - empty - delete
SUPRESSED - delete
JUSTIFICATION - delete
LASTUPDATE - delete
NOTE - delete

```python
del df['DATASOURCE']
df['DAYOFSERVICE'] = df['DAYOFSERVICE'].astype('datetime64[ns]')
df['TRIPID'] = df['TRIPID'].astype('category')
df['PROGRNUMBER'] = df['PROGRNUMBER'].astype('category')
df['STOPPOINTID'] = df['STOPPOINTID'].astype('category')
df['VEHICLEID'] = df['VEHICLEID'].astype('category')

del df['NOTE']
del df['LASTUPDATE']
del df['JUSTIFICATIONID']
del df['SUPPRESSED']
del df['PASSENGERS']
del df['PASSENGERSIN']
del df['PASSENGERSOUT']
del df['DISTANCE']
```
